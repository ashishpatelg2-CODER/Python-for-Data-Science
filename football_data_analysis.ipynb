# Football Data Analysis
This notebook loads football match data from football-data.co.uk and organizes it by league and season.

```
import pandas as pd
```

```
root = 'https://www.football-data.co.uk/mmz4281/'
```

```

leagues = ['E0','E2','E3']
frames = []

for league in leagues:
    df = pd.read_csv(root + "2122/" + league + ".csv")
    frames.append(df)

len(frames)
```

```

dict_countries = {
    'Spanish La Liga':'SP1',
    'Spanish Segunda Division':'SP2',
    'German Bundesliga':'D1',
    'English Premier League':'E0',
    'English League 1':'E2',
    'English League 2':'E3',
}

dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    dict_historical_data[league] = pd.concat(frames)
```
