````python
import os
import subprocess

# Go to your repo
os.chdir(r"D:\web-dev-learning-html") # Use your actual path

def smart_deploy(message):
    # 1. Check for any changes (staged, unstaged, or untracked)
    status = subprocess.check_output(["git", "status", "--porcelain"]).decode().strip()
    
    if not status:
        print("‚ú® Working tree is clean. Nothing to add or commit.")
        return

    try:
        # 2. If changes exist, proceed with deployment
        print(f"Changes detected:\n{status}")
        subprocess.run(["git", "add", "."], check=True)
        subprocess.run(["git", "commit", "-m", message], check=True)
        subprocess.run(["git", "push"], check=True)
        print("üöÄ Successfully deployed to GitHub!")
    except subprocess.CalledProcessError as e:
        print(f"‚ùå Deployment failed: {e}")

# Run it
smart_deploy("Jupyter process update")
````

# üìÅ Working with Directories and Files in Python

## 1Ô∏è‚É£ Changing the Working Directory

You can change the directory for your current session using either a built-in magic command (Jupyter) or the standard Python library.

---

### üîπ Magic Commands (Quickest ‚Äì Jupyter Notebooks)

Use `%cd` to change the directory directly within a cell:

```python
%cd "C:/your/path/here"  # Use forward slashes or raw strings r"C:\path"
%pwd                    # Verify current directory
```

---

### üîπ Python `os` Module (Script-Based Approach)

Use the `os.chdir()` function:

```python
import os

os.chdir('/path/to/directory')
```

---

## 2Ô∏è‚É£ Moving (Cutting) Files and Folders

To "cut" a file or folder from one location to another, use `shutil.move()`.
This removes the item from the source and places it in the destination.

---

### üîπ Move a Single File

```python
import shutil

shutil.move('source_folder/file.txt', 'destination_folder/file.txt')
```

---

### üîπ Move an Entire Folder

```python
import shutil

shutil.move('path/to/old_folder', 'path/to/new_location/')
```

---

### üîπ Move Multiple Files with a Loop

To move all files from one folder to another:

```python
import os
import shutil

source = 'path/to/source/'
destination = 'path/to/destination/'

for file_name in os.listdir(source):
    # Construct full file paths
    source_path = os.path.join(source, file_name)
    # Move file to destination
    shutil.move(source_path, destination)
```

# üìÇ Python File & Directory Management Overview

| Command                   | What it Does                        | Typical Use Case                   | ‚ö†Ô∏è Notes                            |
| ------------------------- | ----------------------------------- | ---------------------------------- | ----------------------------------- |
| `os.remove("file.txt")`   | Deletes a single **file**           | Removing logs, temp files, exports | ‚ùå Fails if target is a folder       |
| `os.rmdir("folder")`      | Deletes an **empty folder**         | Cleaning up unused directories     | ‚ùå Errors if folder is not empty     |
| `shutil.rmtree("folder")` | Deletes folder **and all contents** | Full project cleanup / cache reset | ‚ö†Ô∏è Permanent, no undo               |
| `os.getcwd()`             | Returns current working directory   | Debugging file path issues         | Safe                                |
| `os.chdir("path")`        | Changes current working directory   | Switching project contexts         | Affects relative paths              |
| `os.listdir()`            | Lists all items in a directory      | Bulk processing                    | Returns names only (not full paths) |

# 1Ô∏è‚É£ Navigation & Inspection

### üéØ Objective: Locate and explore your current working environment

Understanding where your script is running prevents most file-related errors.

---

## üîé Get Current Path ‚Äî `os.getcwd()`

```python
import os
print(os.getcwd())
```

**What it does:**

* Returns the absolute path of your current working directory.

**Why it matters:**

* Prevents `FileNotFoundError`
* Essential when running scripts from IDEs, Jupyter, or terminals

**Pro Tip:**
Always check this before performing file operations if you're unsure where your script is executing.

---

## üìã List Directory Contents ‚Äî `os.listdir()`

```python
os.listdir()
```

```python

```

**What it does:**

* Returns a list of filenames and subfolders in the specified directory.

**Use cases:**

* Looping through files
* Filtering by extension
* Bulk renaming or deleting

**Important:**

* Returns only names, not full paths.
* Combine with `os.path.join()` for safe file operations.

---

## üìÇ Change Directory ‚Äî `os.chdir()`

```python

```

**What it does:**

* Changes the working directory.

**Why it's useful:**

* Access files in different project folders
* Prepare scripts for Git deployments
* Work with relative paths cleanly

‚ö†Ô∏è After changing directories, all relative paths refer to the new location.

# 4Ô∏è‚É£ Safety & Validation

# üöÄ Advanced File Handling with `pathlib` (Modern Python Approach)

`pathlib` is the modern, object-oriented alternative to `os` and `os.path`.
It makes file system operations **cleaner, safer, and more readable**.

> Available in Python 3.4+ (fully mature in 3.6+)

---

# üìå Why Use `pathlib`?

‚úÖ Object-oriented paths
‚úÖ Cross-platform compatibility (Windows, macOS, Linux)
‚úÖ Cleaner syntax
‚úÖ Built-in methods for common operations
‚úÖ Reduces string-based path bugs

---

# üèó Getting Started

```python
from pathlib import Path
```

Create a path object:

```python
path = Path("data/report.csv")
```

Unlike `os.path`, this is now a **Path object**, not a string.

# üß† `pathlib` vs `os` Comparison

| Create folder   | `os.mkdir()`       | `Path.mkdir()`   |
| List files      | `os.listdir()`     | `Path.iterdir()` |
